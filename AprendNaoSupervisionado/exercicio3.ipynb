```python
# Aprendizado não supervisionado
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dia_semana = [1,2,3,4,5,6,7]
n_clientes= [15,20,30,100,350,500,700]
dadosDias = np.column_stack((dia_semana,n_clientes))
dadosDias
```

```python
# algoritmo kmeans
kmeans = KMeans(n_clusters=4) 
kmeans.fit(dadosDias)
```

```python
# avaliar o desempenho do aldoritmo kmeans
sse = kmeans.inertia_
print("sse: ", sse)
```

```python
# visualização dos cluters
labels = kmeans.labels_ 
centroids = kmeans.cluster_centers_ 
```

```python
plt.scatter(dadosDias[:,0], dadosDias[:,1], c=labels)
plt.title('Agrupamento dos dias de maior movimento')
plt.xlabel('Dias')
plt.ylabel('Número de clientes')
```
